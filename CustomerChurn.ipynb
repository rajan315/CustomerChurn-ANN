```python
df1['TotalCharges']= pd.to_numeric(df1.TotalCharges,errors='coerce')
```

```python
def print_unique(df):
  for column in df1:
    if df1[column].dtypes=='object':
      print(f'{column} : {df1[column].unique()}')
```

```python
print_unique(df1)
```

```python
df1.replace('No internet service','No',inplace=True)
df1.replace('No phone service','No',inplace=True)
```

```python
print_unique(df1)
```

```python
yes_no_col = ['Partner','Dependents','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','PaperlessBilling',
              'Churn','gender','PhoneService']
for col in yes_no_col:
  df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
for column in df1:
  print(f'{column} : {df1[column].unique()}')
```

```python
df1['gender'] = df1['gender'].replace({'Female':1,'Male':0})
```

```python


df1['InternetService'] = df1.InternetService.replace({'DSL': 0,'Fiber optic': 1,'No': 2})
df1['Contract'] = df1.Contract.replace({'Month-to-month': 0,'One year':1,'Two year': 2})
df1['PaymentMethod'] = df1.PaymentMethod.replace({'Electronic check':0, 'Mailed check':1 ,'Bank transfer (automatic)':2, 'Credit card (automatic)' : 3})
df1.head()
```

```python
df1.info()
```

```python
cols = ['tenure','MonthlyCharges','TotalCharges']
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df1[cols] = scaler.fit_transform(df1[cols])
```

```python

df1.head()
```

```python
X = df1.drop('Churn',axis=1)
y = df1['Churn']
```

```python
y
```

```python
df1.shape
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
model = keras.Sequential([
    keras.layers.Dense(1000, input_shape=(19,),activation='relu'),
    keras.layers.Dense(500,activation='relu'),
    keras.layers.Dense(250,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train,y_train,epochs=20)
```

```python
model.evaluate(X_test,y_test)
```
